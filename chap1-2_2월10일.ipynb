Now, we move to a slightly more advanced analysis. We will look at the attendance of all tracks as a function of the weekday. We can get the weekday easily with pandas: the index attribute of the DataFrame object contains the dates of all rows in the table. This index has a few date-related attributes, including weekday:

```python
df.index.weekday
```

```python
days = np.array(['Monday', 'Tuesday', 'Wednesday', 
                         'Thursday', 'Friday', 'Saturday', 
                         'Sunday'])
df['Weekday'] = days[df.index.weekday]
```

```python
df_week = df.groupby('Weekday').sum()
```

```python
df_week 
```

```python
df_week.plot(lw=3)
plt.ylim(0);  # Set the bottom axis to 0.
```

```python
df_week.ix[days].plot(lw=3)
plt.ylim(0);  # Set the bottom axis to 0.
```
